### 2.3 - Criando Indices

```python
df = df_pixels.sample(frac=1).reset_index(drop=True)
df
```

```python
df.dtypes
```

```python
df['R'] = df['R'].astype('int64')
df['G'] = df['G'].astype('int64')
df['B'] = df['B'].astype('int64')
```

```python
df.dtypes
```

```python
df['Media'] = ((df['R'] + df['G'] + df['B'])/3).round().astype(int)
df
```

```python
df['ExG'] = ((2*df['G']) - df['R'] - df['B']).round()
df
```

```python
df['GLI'] = (((2*df['G']) - df['R'] - df['B'])/(2*df['G']) + df['R'] + df['B']).round().astype(int)
df
```

```python
df['VARI'] = ((df['G'] - df['R'])/(df['G'] + df['R'] - df['B']))
df
```

```python
#Y = 0.299 * R + 0.587 * G + 0.114 * B
df['Gray'] = (0.299*df['R']) + (0.587*df['G']) + (0.114*df['B'])
df
```

## 3 - Rodando Algoritmos de IA Base
RF = Acurcia: %84.71

NB = Acurcia: %80.42

SVM = Acurcia: %84.11


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
x = df[['R','G','B','Media','ExG','GLI','VARI','Gray']]
y = df['Daninha']

SEED = 158020
np.random.seed(SEED)
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, stratify=y)
print('Treinaremos com %d elementos e testaremos com %d elementos'% (len(X_train), len(X_test)))
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=SEED)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurcia: {accuracy}")
```

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred = nb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurcia: {accuracy}")
```

## 6 - Validando Modelos

### 6.1 - As mtricas

```python
#matriz de confuso
from sklearn.metrics import accuracy_score #acuracia

def m_matriz(y_test, y_pred):
  print(pd.crosstab(y_test, y_pred, rownames=['Real'], colnames=['Predito'], margins=True))
```

```python
# Acuracy
from sklearn.metrics import accuracy_score # importando a biblioteca para calcular a acurcia

def m_acuracia(y_test, y_pred):
  acuracia = accuracy_score(y_test, y_pred)
  print('Acurcia: %f' % acuracia)
```

Precision - The best value is 1 and the worst value is 0

```python
# Precision
from sklearn.metrics import precision_score #precision

def m_precision(y_test, y_pred):
  precision = precision_score(y_test, y_pred)
  print('Precision: %f' % precision)
```

Recall - The best value is 1 and the worst value is 0

```python
#Recall
from sklearn.metrics import recall_score #recall

def m_recall(y_test, y_pred):
  recall = recall_score(y_test, y_pred)
  print('Recall: %f' % recall)
```

F1 score - The best value at 1 and worst score at 0

```python
# F1-Score
from sklearn.metrics import f1_score #f1-score

def m_f1(y_test, y_pred):
  f1 = f1_score(y_test, y_pred)
  print('F1-Score: %f' % f1)
```

```python
def processa_imagem(imagem):
    altura, largura, _ = imagem.shape
    tamanho_bloco = 15

    for x in range(0, largura, tamanho_bloco):
        for y in range(0, altura, tamanho_bloco):
            bloco = obtem_bloco(imagem, x, y, tamanho_bloco)
            percentual_branco = calcula_percentual_branco(bloco)

            if percentual_branco > 0.06:
                x+=2
                y+=2
                pinta_bloco(imagem, x, y, tamanho_bloco)

    cv2_imshow(imagem)
    return(imagem)

def obtem_bloco(imagem, x, y, tamanho_bloco):
    bloco = imagem[y:y+tamanho_bloco, x:x+tamanho_bloco]
    return bloco

def calcula_percentual_branco(bloco):
    total_pixels = bloco.size
    brancos = np.count_nonzero((bloco == [255, 255, 255]).all(axis=2))  # Conta pixels brancos em uma imagem BGR

    return brancos / total_pixels

def pinta_bloco(imagem, x, y, tamanho_bloco):
    imagem[y:y+tamanho_bloco, x:x+tamanho_bloco] = [0, 0, 255]  # Pinta o bloco de vermelho (BGR)
```

```python
def limpa_img(imagem, base):
  altura, largura, _ = imagem.shape

  for x in range(0, largura):
    for y in range(0, altura):
      g = imagem[x,y,0]
      b = imagem[x,y,1]
      r = imagem[x,y,2]

      if ((r>=253) & (g>=253) & (b>=253)):
        imagem[x,y,0] = base[x,y,0]
        imagem[x,y,1] = base[x,y,1]
        imagem[x,y,2] = base[x,y,2]

  return(imagem)
```
